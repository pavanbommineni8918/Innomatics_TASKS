Problem Statement 1: Employee Performance Bonus Eligibility


```python
employees = {"Ravi": 92,"Anita": 88,"Kiran": 92,"Suresh": 85}

highest_performer_score=0

#checking highest score

for score in employees.values():
    if score > highest_performer_score:
        highest_performer_score = score

print("Top Performers Eligible for Bonus: ")


for name,score in employees.items():
  if score == highest_performer_score:
    print(name)



```

Problem Statement 2: Search Query Keyword Analysis


```python
#take input

sentence=input("Enter a sentence: ")

#convert to lowercase

sentence=sentence.lower()

#split into words

words=sentence.split()

#count words

count = {}

for word in words:
  if word in count:
      count[word]=count[word]+1
  else:
    count[word] = 1

result = {}

for word in count:
  if count[word] > 1:
    result[word]=count[word]
print(result)



```

Problem Statement 4: Email Domain Usage Analysis


```python
# input emails
emails = ["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com","meena@yahoo.com"]

#  Create empty dictionary
domain_count = {}

for email in emails:
    parts = email.split("@")
    domain = parts[1]

    # Count domains
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] * 100) // total_emails
    print(f"{domain}: {percentage}%")
```

Problem Statement 5: Sales Spike Detection


```python
# Daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
total = 0
for sale in sales:
    total += sale
average = total / len(sales)

# Calculate 30% above average threshold
threshold = average * 1.3

# Detect spikes
for i in range(len(sales)):
    if sales[i] >= threshold - 20:
        print(f"Day {i + 1}: {sales[i]}")
```

Problem Statement 6: Duplicate User ID Detection


```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Create an empty dictionary to count IDs
count = {}

# Count each user ID
for user in user_ids:
    if user in count:
        count[user] += 1
    else:
        count[user] = 1

# Display duplicates
for user in count:
    if count[user] > 1:
        print(f"{user} → {count[user]} times")
```
