```python
# Task 4: Chat Prompt System (Improved & Original Version)

from langchain_core.prompts import ChatPromptTemplate

def build_chat_prompt(user_role, topic_name):
    """
    Creates a chat-based prompt using role-based behavior.
    """

    # Define role behaviors
    role_behaviors = {
        "teacher": "You explain concepts in a simple and structured way.",
        "interviewer": "You ask thoughtful and challenging questions.",
        "motivator": "You inspire and encourage learning with positivity."
    }

    # Select behavior (default fallback included)
    system_behavior = role_behaviors.get(
        user_role,
        "You are a helpful AI assistant."
    )

    # Create chat template
    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_behavior),
        ("user", "Discuss the topic: {topic}")
    ])

    # Generate formatted messages
    return chat_prompt.format_messages(topic=topic_name)


# ✅ Test Example
result = build_chat_prompt("teacher", "Neural Networks")

for msg in result:
    print(msg)
```

```python
# Task 5: Input Validation System

def check_user_inputs(audience_level, tone_style):
    """
    Validates audience and tone inputs.
    If invalid, assigns safe default values.
    """

    allowed_audience = {"beginner", "intermediate", "expert"}
    allowed_tones = {"formal", "casual", "fun"}

    # Validate audience
    if audience_level not in allowed_audience:
        print("⚠️ Invalid audience! Defaulting to 'beginner'")
        audience_level = "beginner"

    # Validate tone
    if tone_style not in allowed_tones:
        print("⚠️ Invalid tone! Defaulting to 'formal'")
        tone_style = "formal"

    return audience_level, tone_style


# 🔹 Test
aud, tone = check_user_inputs("kids", "cool")
print(aud, tone)
```

```python
# Task 6: Dynamic Prompt Generator

from langchain_core.prompts import PromptTemplate

def create_prompt(topic_name, audience_level, tone_style, format_type):
    """
    Generates prompts dynamically based on style type.
    """

    # Step 1: Validate inputs
    audience_level, tone_style = check_user_inputs(audience_level, tone_style)

    # Step 2: Define templates dictionary
    prompt_styles = {
        "teaching": "Explain {topic} to {audience} in a {tone} teaching manner",
        "interview": "Generate interview questions on {topic} for {audience} with a {tone} tone",
        "storytelling": "Describe {topic} as a story for {audience} in a {tone} style"
    }

    # Step 3: Handle invalid style
    if format_type not in prompt_styles:
        return "Style not recognized"

    # Step 4: Create PromptTemplate dynamically
    dynamic_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=prompt_styles[format_type]
    )

    # Step 5: Return formatted prompt
    return dynamic_template.format(
        topic=topic_name,
        audience=audience_level,
        tone=tone_style
    )


# 🔹 Test
print(create_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# Task 7: Reusability Demonstration

from langchain_core.prompts import PromptTemplate

# Single reusable template
common_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Provide an explanation of {topic} suitable for {audience} in a {tone} tone"
)

# Different input combinations
test_inputs = [
    ("Artificial Intelligence", "beginner", "formal"),
    ("Cybersecurity", "expert", "formal"),
    ("Python Programming", "beginner", "fun"),
    ("Data Science", "intermediate", "casual"),
    ("Cloud Computing", "expert", "formal")
]

# Execute tests
for topic, audience, tone in test_inputs:
    output = common_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
    print(output)
```
