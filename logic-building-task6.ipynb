## Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
     # Initialize parked vehicle count
    current_parked = 0
    peak_usage = 0
    
    # Loop through vehicle logs
    for log in vehicle_logs:

        if log == 'IN':
            current_parked += 1 # Increase count when vehicle enters
        elif log == 'OUT':
            current_parked -= 1 # Decrease count when vehicle exits
         # Track peak parking usage
        if current_parked > peak_usage:
            peak_usage = current_parked
    #check parking capacity
    if current_parked > capacity:
        status = 'Alert! Parking Exceeded Capacity'
    else:
        status = 'Available'
    #display results
    print('Currently Parked Vehicles:', current_parked)
    print('Peak Parking Usage:', peak_usage)
    print('Parking Status:', status)
#input data
smart_parking_system(50, ['IN','IN','IN','OUT','IN','IN','OUT'])
```

## Problem 2: Online Food Delivery Time Estimator

```python
def delivery_time_estimator(distance, traffic, weather):
    # Base time calculation (5 minutes per km)
    base_time = distance * 5

    # Traffic impact
    if traffic == "High":
        base_time += 15
    elif traffic == "Medium":
        base_time += 10
    else:
        base_time += 5

    # Weather impact
    if weather == "Rainy":
        base_time += 10
    elif weather == "Storm":
        base_time += 20

    # Final ETA
    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
delivery_time_estimator(8, "High", "Rainy")
```

## Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def theatre_seat_analyzer(total_seats, booked_seats):
    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    # Suggest additional shows
    if occupancy >= 90:
        suggestion = "Consider Opening Additional Shows"
    else:
        suggestion = "No Need for Extra Show"

    # Display results
    print("Occupancy:", occupancy, "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)


# Sample Input
theatre_seat_analyzer(200, [1]*150)
```

## Problem 4: Cloud Server Load Classification System

```python
def server_load_classifier(cpu_readings):
    # Calculate average CPU usage
    total = 0
    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    # Classify server load
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    # Display results
    print("Average CPU Load:", average_cpu, "%")
    print("Server Status:", status)


# Sample Input
server_load_classifier([45, 60, 70, 85, 90])
```

## Problem 5: Smart Classroom Resource Usage Monitor

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Define threshold (example: 8 hours)
    threshold = 8

    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > threshold:
            overused.append(resource)

    # Energy alert condition
    if len(overused) > 0:
        alert = "Yes"
    else:
        alert = "No"

    # Display results
    print("Overused Resources:", ", ".join(overused))
    print("Energy Alert:", alert)


# Sample Input
classroom_resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

## Problem 6: Online Event Registration Capacity Controller

```python
def event_registration_controller(capacity, registrations):
    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    # Display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration_controller(100, 105)
```
