## Problem 1: Smart Light Controller

Create class for smart light

```python
class SmartLight:
    def __init__(self, light_name, status):
        self.light_name = light_name
        self.status = status
    
    def status_display(self):
        print(f"{self.light_name} is {self.status}")

smart_light = SmartLight("Bedroom Light", "ON")
smart_light.status_display()
```

## Problem 2: Employee ID Card System

Create class to store employee details and displays their details

```python
class EmployeeDetails:
    def __init__(self, employee_name, employee_ID, department):
        self.employee_name = employee_name
        self.employee_ID = employee_ID
        self.department = department

    def display_ID(self):
        print("Employee ID Card")
        print(f"Name: {self.employee_name}")
        print(f"ID: {self.employee_ID}")
        print(f"Department: {self.department}")

e = EmployeeDetails("Rahul", "EMP102", "IT")
e.display_ID()
```

## Problem 3 : Mobile Contact Record

Create class to contact details and dispay them

```python
class ContactDetails:
    def __init__(self, name, number):
        self.name = name
        self.number = number
    
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.number}")
    
m = ContactDetails("Anita", 9876543210)
m.display_contact()
```

## Problem 4: Product Price Tag Generator

Create class to store product name and price and display them

```python
class ProductInfo:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def display_product(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

p = ProductInfo("Headphones", 2499)
p.display_product()
```

## Problem 5: Movie Rating Display System

Create class to store movie name and rating and display them

```python
class MovieInfo:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    def display_movie(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating}/5")

mo = MovieInfo("Inception", 4.8)
mo.display_movie()
```

## Problem 6: Delivery Address Manager

Create class to store customer name and address and display them

```python
class CustomerInfo:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_customer(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
    
d = CustomerInfo("Suman", "Hyderabad")
d.display_customer()
```
