```python

# Problem 1: Social Media - Post Engagement Evaluator

def evaluate_post_engagement(post_likes):
    # Calculate total likes using dictionary values
    total_likes = sum(post_likes.values())
    print("Overall Likes:", total_likes)

    # Check if post is viral or normal
    if total_likes >= 1000:
        return "Post Result: Viral Post"
    else:
        return "Post Result: Normal Engagement"

posts = {"postA": 250, "postB": 300, "postC": 400, "postD": 500}
print(evaluate_post_engagement(posts))
```

```python

# Problem 2: Healthcare - Medicine Inventory Monitor

def check_medicine_inventory(stock_data):
    # Create dictionary to store report
    status_report = {}

    # Iterate through medicines
    for name, quantity in stock_data.items():
        # Check stock level
        if quantity < 10:
            condition = "Reorder Required"
        else:
            condition = "Stock Available"

        print(f"{name} -> Quantity: {quantity} | Status: {condition}")
        status_report[name] = condition

    return status_report

medicine_stock = {"Paracetamol": 15, "Insulin": 12, "Antibiotic": 9}
check_medicine_inventory(medicine_stock)
```

```python

# Problem 3: Agriculture - Rainfall Analysis System

def analyze_rainfall(data):
    # Minimum required rainfall level
    minimum_requirement = 55

    # Calculate average rainfall
    avg_rainfall = sum(data.values()) / len(data)
    print("Average Rainfall:", avg_rainfall)

    # Compare with required level
    if avg_rainfall >= minimum_requirement:
        return "Rainfall Condition: Sufficient"
    else:
        return "Rainfall Condition: Insufficient"

rain_data = {"Monsoon": 100, "Winter": 50, "Summer": 60}
print(analyze_rainfall(rain_data))
```

```python

# Problem 4: Social Media - Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    # Dictionary to count occurrences
    count_dict = {}

    # Count each username
    for user in usernames:
        count_dict[user] = count_dict.get(user, 0) + 1

    # Check for duplicates
    for user, count in count_dict.items():
        if count > 1:
            return "Duplicate Accounts: Detected"

    return "Duplicate Accounts: Not Found"

accounts_list = ["aruncode_12", "ishan_creation", "aruncode_12"]
print(detect_duplicate_accounts(accounts_list))
```

```python

# Problem 5: Healthcare - Appointment Eligibility Checker

def verify_appointment_eligibility(name, age):
    # Check age criteria
    if age >= 18:
        print(f"Patient age:{age}")
        return f"{name.title()} is Eligible for Appointment"
    else:
        return f"{name.title()} is Not Eligible for Appointment"

patient_name = input("Enter patient name: ")
patient_age = int(input("Enter patient age: "))
print(verify_appointment_eligibility(patient_name, patient_age))
```

```python

# Problem 6: Agriculture - High Value Crop Filter

def filter_high_value_crops():
    premium_list = []

    # Ask number of crops
    total_crops = int(input("How many crops to evaluate? "))

    for i in range(total_crops):
        crop_name = input("Enter crop name: ")
        crop_price = int(input("Enter crop price: "))

        # Check if price greater than 2000
        if crop_price > 2000:
            premium_list.append(crop_price)

    return f"Premium Crops: {premium_list}"

print(filter_high_value_crops())
```

```python

# Problem 7: System Monitoring - Error Health Analyzer

def system_health_report(status_list):
    # Count number of errors
    error_count = status_list.count("Error")
    print("Total Errors:", error_count)

    # Determine system health
    if error_count == 0:
        print("System Healthy")
    elif error_count <= 5:
        print("Minor Issues Detected")
    else:
        print("Critical Condition")

health_logs = ["Error", "NoError", "Error", "Error", "NoError"]
system_health_report(health_logs)
```

```python

# Problem 8: Banking - Daily Transaction Validator

def verify_daily_transactions():
    transaction_limit = 50000
    total_spent = 0

    # Ask number of transactions
    transactions = int(input("Number of transactions today: "))

    # Add each transaction amount
    for i in range(transactions):
        amount = int(input(f"Enter amount for Transaction {i+1}: "))
        total_spent += amount

    print("Total Transaction Amount:", total_spent)

    # Check against daily limit
    if total_spent <= transaction_limit:
        return "Transaction Approved"
    else:
        return "Transaction Declined"

print(verify_daily_transactions())
```

```python

# Problem 9: E-Learning - Attendance Eligibility Checker

def calculate_attendance():
    days_present = int(input("Days Present: "))
    days_absent = int(input("Days Absent: "))

    total_days = days_present + days_absent

    # Avoid division by zero
    if total_days == 0:
        return "No attendance data available."

    attendance_percent = (days_present / total_days) * 100
    print("Attendance Percentage:", attendance_percent)

    # Check eligibility
    if attendance_percent >= 75:
        return "Exam Eligibility: Approved"
    else:
        return "Exam Eligibility: Denied"

print(calculate_attendance())
```

```python

# Problem 10: Smart Electricity Bill Calculator

def calculate_electricity_bill(units_consumed):
    # Apply slab-wise billing rules
    if units_consumed <= 100:
        total = units_consumed * 3
    elif units_consumed <= 200:
        total = (100 * 3) + (units_consumed - 100) * 5
    else:
        total = (100 * 3) + (100 * 5) + (units_consumed - 200) * 7

    return total

units = int(input("Enter total electricity units consumed: "))
bill_amount = calculate_electricity_bill(units)

print("Total Bill Amount:", bill_amount)

# Categorize usage
if bill_amount < 500:
    print("Usage Category: Low")
elif bill_amount <= 1500:
    print("Usage Category: Moderate")
else:
    print("Usage Category: High")
```
